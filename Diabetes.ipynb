```python
import pandas as pd
import numpy as np
```

```python
## reading the features and the labels
data = pd.read_csv(r'C:\Users\prama\Desktop\Afitech notes\Bagging_RandomForest\diabetes.csv')
```

```python
cols =[ 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age' ]
```

```python
# as mentioned in the data description, the missing values have been replaced by zeroes. So, we are replacing zeroes with nan
for col in cols:
    data[col] =  data[col].replace(0,np.nan)
```

```python
# checking for missing values
data.isna().sum()
```

```python
# imputing the missing values
data['Glucose'] = data['Glucose'].fillna(data['Glucose'].mode()[0])
data['BloodPressure'] = data['BloodPressure'].fillna(data['BloodPressure'].mode()[0])
data['SkinThickness'] = data['SkinThickness'].fillna(data['SkinThickness'].mean())
data['Insulin'] = data['Insulin'].fillna(data['Insulin'].mean())
data['BMI'] = data['BMI'].fillna(data['BMI'].mean())
```

```python
# checking for missing values after imputation
data.isna().sum()
```

```python
#Separating the feature and the Label columns 
x=data.drop(labels='Outcome', axis=1)
y= data['Outcome']
```

```python
x.sample(n=6)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# as the datapoints differ a lot in magnitude, we'll scale them
scalar = StandardScaler()
x_scaled = scalar.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y, test_size= 0.25)
```

```python
import xgboost as xgb
import pickle
from sklearn import datasets
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
# fit model no training data
model = XGBClassifier(objective='binary:logistic')

model.fit(x_train, y_train)
```

```python
# cheking training accuracy
y_pred = model.predict(x_train)

predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(y_train,predictions)
accuracy
```

```python
# cheking initial test accuracy
y_pred = model.predict(x_test)

predictions = [round(value) for value in y_pred]

accuracy = accuracy_score(y_test,predictions)
accuracy
```

```python
# Now to increae the accuracy of the model, we'll do hyperparameter tuning using grid search 
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
   
    ' learning_rate':[1,0.5,0.1,0.01,0.001],
    'max_depth': [3,5,10,20],
    'n_estimators':[12,50,100,200]
    
}
```

```python
grid= GridSearchCV(XGBClassifier(objective='binary:logistic'),param_grid, verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
# To  find the parameters givingmaximum accuracy
grid.best_params_
```

```python
# Create new model using the same parameters
new_model=XGBClassifier(learning_rate= 1, max_depth= 3, n_estimators= 12)
new_model.fit(x_train, y_train)
```

```python
y_pred_new = new_model.predict(x_test)

predictions_new = [round(value) for value in y_pred_new]

accuracy_new = accuracy_score(y_test,predictions_new)

accuracy_new
```
